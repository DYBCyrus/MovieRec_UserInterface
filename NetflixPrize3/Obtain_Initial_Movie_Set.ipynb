```python
import pandas as pd
import numpy as np
from collections import defaultdict
import random
import json
```

```python
imdb_df = pd.read_csv('IMDB_Final_Movies.csv')
imdb_df.head()
```

```python
print(imdb_df.shape)
imdb_df.dropna(inplace=True)
print(imdb_df.shape)
```

```python
imdb_df.to_csv("IMDB_Final_Movies.csv")
```

```python
genre_to_movie_rating_pair = defaultdict(list)
for i in range(imdb_df.shape[0]):
    title = imdb_df.iloc[i]['primaryTitle']
    year = str(int(imdb_df.iloc[i]['startYear']))
    genres = imdb_df.iloc[i]['genres'].split(',')
    rate = float(imdb_df.iloc[i]['averageRating'])
    for g in genres:
        genre_to_movie_rating_pair[g].append((title,year,rate))
```

```python
sorted_genre_to_movie_rating_pair = defaultdict(list)
for k,v in genre_to_movie_rating_pair.items():
    sorted_genre_to_movie_rating_pair[k] = sorted(v, key=lambda x: x[2], reverse=True)
```

```python
json.dump(sorted_genre_to_movie_rating_pair, open("sorted_movies_for_genres.json",'w'))
```
